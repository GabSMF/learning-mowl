```python
import mowl
from rdflib import Graph
from mowl.datasets.base import PathDataset

mowl.init_jvm("5g")
```

```python
g = Graph()
g.parse("university_ontology.ttl", format="turtle")
g.serialize("university_ontology.owl", format="xml")
```

```python
from mowl.datasets.base import PathDataset
from mowl.corpus import extract_axiom_corpus
from mowl.corpus import extract_annotation_corpus
ds = PathDataset("university_ontology.owl")

ontology = ds.ontology
corpus_axiom = extract_axiom_corpus(ontology)
corpus_annotation = extract_annotation_corpus(ontology)
```

```python
corpus_annotation
```

```python
from mowl.models import SyntacticPlusW2VModel
```

```python
model = SyntacticPlusW2VModel(ds, corpus_filepath="test")
model.set_w2v_model(min_count=1)
model.generate_corpus(save=True, with_annotations=True)
model.train()
```

```python
w2v = model.w2v_model
```

```python
print(w2v.wv.index_to_key[:10])

vector = w2v.wv["https://w3id.org/ontouml/university_ontology#Student"]
print(vector)

sim = w2v.wv.similarity(
    "https://w3id.org/ontouml/myontology#Student",
    "https://w3id.org/ontouml/myontology#RegularStudent"
)
print("Similaridade:", sim)
```
