```python
import mowl
from mowl.datasets.builtin import PPIYeastSlimDataset
from mowl.corpus import extract_axiom_corpus, extract_annotation_corpus
from mowl.models import SyntacticPlusW2VModel
from mowl.datasets import PathDataset
mowl.init_jvm("5g")
```

```python
ds = PathDataset("university_ontology.owl")
train_ontology = ds.ontology
```

```python
axiom_corpus = extract_axiom_corpus(ds.ontology)
annotation_corpus = extract_annotation_corpus(ds.ontology)
```

```python
model = SyntacticPlusW2VModel(ds, corpus_filepath="test2")
model.set_w2v_model(min_count=1)
model.generate_corpus(save=True, with_annotations=True)
model.train()
```

```python
class_embeds = model.class_embeddings
class_embeds
```

```python
# Embeddings de propriedades de objeto OWL
obj_prop_embeds = model.object_property_embeddings
obj_prop_embeds
```

```python
ind_embeds = model.individual_embeddings
ind_embeds
```

```python
from mowl.visualization import TSNE

name_to_embedding = class_embeds
name_to_label = {name: "Class" for name in class_embeds.keys()}

tsne = TSNE(name_to_embedding, name_to_label)
tsne.generate_points(250, workers=4)
```
